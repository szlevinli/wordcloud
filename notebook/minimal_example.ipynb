```python
import os
from os import path
from PIL import Image
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# get current work directory
d = os.getcwd()
# get data directory
data_dir = path.join(d, '..', 'data')
# get mask directory
mask_dir = path.join(d, '..', 'mask')
# get output directory
out_dir = path.join(d, '..', 'out')
# get fon directory
font_dir = path.join(d, '..', 'font')
```

```python
# 设置字体 用于显示中文
font = path.join(font_dir, 'hyxdlb.ttf')
# get file name
file_name = path.join(data_dir, 'intelligent_trend.txt')
# background img
mask = np.array(Image.open(path.join(mask_dir, 'cloud.png')))
# 读取文件
with open(file_name, 'r', encoding='UTF-8') as f:
    data = f.read()
    # 生成词云
    wc = WordCloud(collocations=False, font_path=font, mask=mask,
                   width=800, height=600, margin=2,
                   background_color='white').generate(data)
    # 显示词云
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    # 存储词云到文件
    wc.to_file(path.join(out_dir, 'intelligent_trend.png'))
```
